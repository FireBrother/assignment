```python
def dump_data(api, start_year=2005, end_year=2018):
    print(api.__name__, start_year, end_year)
    data = []
    for year in range(start_year, end_year+1):
        for quarter in range(1, 5):
            print(year, quarter)
            try:
                df = api(year, quarter)
            except OSError:
                return data
            df['year'] = year
            df['quarter'] = quarter
            data.append(df)
    data = pd.concat(data)
    data = data.reset_index(drop=True)
    return data
```

```python
data_params = [(ts.get_report_data, 'report_data'), (ts.get_profit_data, 'profit_data'),
               (ts.get_operation_data, 'operation_data'), (ts.get_growth_data, 'growth_data'),
               (ts.get_debtpaying_data, 'debtpaying_data'), (ts.get_cashflow_data, 'cashflow_data')]
for param in data_params:
    df = dump_data(param[0], 2005, 2018)
    df.to_hdf('storage.h5', param[1])
```
